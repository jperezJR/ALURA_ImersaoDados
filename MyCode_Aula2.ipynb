```python
#Importar lib Pandas e ler arquivo csv de um Zip file

import pandas as pd

dados = pd.read_csv('https://github.com/jperezJR/ALURA_ImersaoDados/blob/main/dados_experimentos.zip?raw=true', compression= 'zip')
```

```python
#imprimir as 5 primeiras linhas da tabela

dados.head()
```

```python
#renomear coluna index 4 de "droga" para 'composto'

mapa = {'droga' : 'composto'}

dados.rename(columns = mapa)
```

```python
dados.head()
```

```python
# para a mudanca de nomes da coluna de 'droga' para 'composto', devo setar inplace = True como parametro do metodo rename()

dados.rename(columns = mapa, inplace = True)
dados.head()
```

```python
#plotar garfico de barras utilizando a biblioteca seaborn

import seaborn as sns
import matplotlib.pyplot as plt


sns.set()

sns.countplot(dados['tempo'])
```

```python
## ou da forma abaixo, passando o nome do eixo X ele sabera qual coluna pegar do DataFrame dados

sns.countplot(x = 'tempo', data = dados)
```

```python
# plotar grafico da coluna composto. O problema eh que existem muitos valores para essa coluna. Uma forma de melhor avaliar, seria analisar os 5 compostos com maior frequencia
sns.countplot(x = 'composto', data = dados)
```

```python
#descobrir quais compostos apresentam maior frequencia de uso e selecionar os 5 primeiros

dados['composto'].value_counts()

```

```python
cod_compostos = dados['composto'].value_counts().index[:5]

cod_compostos
```

```python
# filtrar tabela para que mostre somente resultados utilizando os 5 mais frequentes compostos utilizados

dados_freqCompostos = dados.query('composto in @cod_compostos')

dados_freqCompostos



```

```python
#plotar grafico dos compostos mais utilizados

sns.countplot(x = 'composto', data = dados_freqCompostos)
```

```python
#utilizando diferentes estilos

sns.countplot(x = 'composto', data = dados_freqCompostos, palette = 'flare')
```

```python
#utilizando diferentes estilos e utilizando a lib matplotlib para mais funcionalidades

plt.figure(figsize=(10, 8))
sns.set_theme(context='paper', style='whitegrid')

ax = sns.countplot(x = 'composto', data = dados_freqCompostos, order = cod_compostos, palette = 'mako')
ax.set_title("Top 5 Compostos",fontsize=20, weight = "bold")
ax.set_xlabel('Composto', fontsize = 12)
ax.set_ylabel('FrequÃªncia', fontsize = 12)



plt.show()
```

```python
#utilizando diferentes estilos e utilizando a lib matplotlib para mais funcionalidades

plt.figure(figsize=(13, 8))
sns.set_theme(context='paper', style='whitegrid')

ax = sns.countplot(x = 'composto', data = dados_freqCompostos, order = cod_compostos, palette = 'mako')
ax.set_title("Top 5 Compostos",fontsize=20, weight = "bold")
ax.set_xlabel('Composto', fontsize = 12)
ax.set_ylabel('FrequÃªncia', fontsize = 12)
posX = 0
for posY in dados['composto'].value_counts().values[:5]:
  ax.annotate(posY, xy = (posX,posY), horizontalalignment = 'center', rotation=45)
  posX += 1
  
plt.show()
```
